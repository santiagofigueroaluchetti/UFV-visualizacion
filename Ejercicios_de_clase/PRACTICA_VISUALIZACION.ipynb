```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('datos_ejercicio_ventas.csv')
print(df.head())
df
```

#**1. Distribucion de ventas por:**

## País

```python
# Agrupar por COUNTRY y sumar el AMOUNT para cada país
ventas_por_pais = df.groupby('COUNTRY')['AMOUNT'].sum()

# Mostrar la distribución de ventas por país
print(ventas_por_pais)

# Agrupar por COUNTRY y sumar el AMOUNT
ventas_por_pais = df.groupby('COUNTRY')['AMOUNT'].sum().reset_index()

# Crear el gráfico de barras interactivo
fig_pais = px.bar(ventas_por_pais, x='COUNTRY', y='AMOUNT', title='Distribución de Ventas por País',
                  labels={'COUNTRY': 'País', 'AMOUNT': 'Ventas (AMOUNT)'})
fig_pais.show()
```

## Año y mes

```python
ventas_por_mes_ano = df.groupby(['YEAR', 'MONTH'])['AMOUNT'].sum().unstack()
print(ventas_por_mes_ano)
# Agrupar por YEAR y MONTH y sumar el AMOUNT
ventas_por_mes_ano = df.groupby(['YEAR', 'MONTH'])['AMOUNT'].sum().reset_index()

# Crear el gráfico de barras interactivo
fig_mes_ano = px.bar(ventas_por_mes_ano, x='MONTH', y='AMOUNT', color='YEAR', barmode='group',
                     title='Distribución de Ventas por Mes y Año',
                     labels={'MONTH': 'Mes', 'AMOUNT': 'Ventas (AMOUNT)', 'YEAR': 'Año'})
fig_mes_ano.show()


```

## Marca

```python
ventas_por_marca = df.groupby('SUBBRAND')['AMOUNT'].sum()
print(ventas_por_marca)

import plotly.express as px

# Agrupar por SUBBRAND y sumar el AMOUNT
ventas_por_marca = df.groupby('SUBBRAND')['AMOUNT'].sum().reset_index()

# Crear el gráfico de barras interactivo
fig = px.bar(ventas_por_marca, x='SUBBRAND', y='AMOUNT', title='Distribución de Ventas por Submarca',
             labels={'SUBBRAND': 'Submarca', 'AMOUNT': 'Ventas (AMOUNT)'})

# Mostrar el gráfico
fig.show()
```

# **2. Tendencia y estacionalidad**

## Todas las ventas del país con menos ventas

```python
import plotly.graph_objects as go
import plotly.express as px

# Encontrar el país con menos ventas
pais_menos_ventas = df.groupby('COUNTRY')['AMOUNT'].sum().idxmin()
print(f"País con menos ventas: {pais_menos_ventas}")

# Filtrar los datos para el país con menos ventas
df_pais_menos_ventas = df[df['COUNTRY'] == pais_menos_ventas]

# Agrupar por año y mes para obtener una serie temporal de ventas mensuales
df_pais_menos_ventas['Fecha'] = pd.to_datetime(df_pais_menos_ventas[['YEAR', 'MONTH']].assign(DAY=1))
ventas_mensuales = df_pais_menos_ventas.groupby('Fecha')['AMOUNT'].sum()

# Crear el gráfico interactivo con Plotly
fig = go.Figure()

# Ventas Mensuales
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=ventas_mensuales, mode='lines', name='Ventas Mensuales'))

# Descomponer manualmente la tendencia y estacionalidad
trend = ventas_mensuales.rolling(window=12, center=True).mean()  # Tendencia simple con promedio móvil
seasonal = ventas_mensuales - trend  # Estacionalidad aproximada

# Agregar la tendencia al gráfico
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=trend, mode='lines', name='Tendencia'))

# Agregar la estacionalidad al gráfico
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=seasonal, mode='lines', name='Estacionalidad'))

# Agregar los residuos al gráfico
residual = ventas_mensuales - (trend + seasonal)
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=residual, mode='lines', name='Residuo'))

# Configuración del diseño del gráfico
fig.update_layout(
    title=f"Descomposición de Ventas para el País con Menos Ventas: {pais_menos_ventas}",
    xaxis_title="Fecha",
    yaxis_title="Ventas (AMOUNT)",
    height=800,
)

# Mostrar el gráfico
fig.show()
```

## Marca con más ventas

```python
import plotly.graph_objects as go
import plotly.express as px

# Encontrar la submarca con más ventas
submarca_mas_ventas = df.groupby('SUBBRAND')['AMOUNT'].sum().idxmax()
print(f"Submarca con más ventas: {submarca_mas_ventas}")

# Filtrar los datos para la submarca con más ventas
df_submarca_mas_ventas = df[df['SUBBRAND'] == submarca_mas_ventas]

# Agrupar por año y mes para obtener una serie temporal de ventas mensuales
df_submarca_mas_ventas['Fecha'] = pd.to_datetime(df_submarca_mas_ventas[['YEAR', 'MONTH']].assign(DAY=1))
ventas_mensuales = df_submarca_mas_ventas.groupby('Fecha')['AMOUNT'].sum()

# Crear el gráfico interactivo con Plotly
fig = go.Figure()

# Ventas Mensuales
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=ventas_mensuales, mode='lines', name='Ventas Mensuales'))

# Descomponer manualmente la tendencia y estacionalidad
trend = ventas_mensuales.rolling(window=12, center=True).mean()  # Tendencia simple con promedio móvil
seasonal = ventas_mensuales - trend  # Estacionalidad aproximada

# Agregar la tendencia al gráfico
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=trend, mode='lines', name='Tendencia'))

# Agregar la estacionalidad al gráfico
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=seasonal, mode='lines', name='Estacionalidad'))

# Agregar los residuos al gráfico
residual = ventas_mensuales - (trend + seasonal)
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=residual, mode='lines', name='Residuo'))

# Configuración del diseño del gráfico
fig.update_layout(
    title=f"Descomposición de Ventas para la Submarca con Más Ventas: {submarca_mas_ventas}",
    xaxis_title="Fecha",
    yaxis_title="Ventas (AMOUNT)",
    height=800,
)

# Mostrar el gráfico
fig.show()
```

# **3. Cuales son las predicciones hechas en España y como de buenas son**



*   AI_forecast  ->  venta que se ha predicho
*   actual  -> venta que se ha producido




```python
import pandas as pd
import plotly.graph_objects as go

# Filtrar los datos para España y para el rango de fechas especificado
data_spain = df[(df['COUNTRY'] == 'Spain') &
                ((df['YEAR'] == 2023) | (df['YEAR'] == 2024)) &
                (df['MONTH'] <= 8 if df['YEAR'].max() == 2024 else True)]

# Crear DataFrames separados para las ventas actuales y predichas
df_actual = data_spain[data_spain['SCENARIO'] == 'actual'].copy()
df_AI_forecast = data_spain[(data_spain['SCENARIO'] == 'AI_forecast') &
                            (data_spain['FORECAST'].isin(['AI_P02F', 'AI_P03F']))].copy()

# Crear la columna de fecha
df_actual['DATE'] = pd.to_datetime(df_actual[['YEAR', 'MONTH']].assign(DAY=1))
df_AI_forecast['DATE'] = pd.to_datetime(df_AI_forecast[['YEAR', 'MONTH']].assign(DAY=1))

# Agrupar y promediar las ventas por mes y año
sales_actual = df_actual.groupby(df_actual['DATE'].dt.to_period("M"))['AMOUNT'].mean().reset_index()
sales_actual['DATE'] = sales_actual['DATE'].dt.to_timestamp()

sales_forecast = df_AI_forecast.groupby(df_AI_forecast['DATE'].dt.to_period("M"))['AMOUNT'].mean().reset_index()
sales_forecast['DATE'] = sales_forecast['DATE'].dt.to_timestamp()

# Unir ambos DataFrames por fecha, manteniendo solo las filas donde hay datos en ambos
comparacion = pd.merge(sales_actual, sales_forecast, on='DATE', how='inner', suffixes=('_Actual', '_Forecast'))
comparacion = comparacion.dropna(subset=['AMOUNT_Forecast'])

# Calcular el MAE entre ventas reales y predichas
comparacion['MAE'] = abs(comparacion['AMOUNT_Actual'] - comparacion['AMOUNT_Forecast'])

# Crear el gráfico con plotly
fig = go.Figure()

# Añadir la línea de ventas reales
fig.add_trace(go.Scatter(
    x=comparacion['DATE'],
    y=comparacion['AMOUNT_Actual'],
    mode='lines+markers',
    name='Ventas Reales (Actual)',
    line=dict(color='blue')
))

# Añadir la línea de ventas predichas
fig.add_trace(go.Scatter(
    x=comparacion['DATE'],
    y=comparacion['AMOUNT_Forecast'],
    mode='lines+markers',
    name='Ventas Predichas (AI Forecast)',
    line=dict(color='red')
))

# Añadir las barras del MAE
fig.add_trace(go.Bar(
    x=comparacion['DATE'],
    y=comparacion['MAE'],
    name=f'MAE: {comparacion["MAE"].mean():.2f}',
    marker_color='gray',
    opacity=0.3
))

# Configuración del diseño del gráfico
fig.update_layout(
    title='Comparación de Ventas Reales vs. Predichas (España)',
    xaxis_title='Fecha',
    yaxis_title='Cantidad Ventas',
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5),
    template="plotly_white"
)

fig.show()
```
