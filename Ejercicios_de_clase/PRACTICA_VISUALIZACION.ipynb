```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('datos_ejercicio_ventas.csv')
print(df.head())
df
```

#**1. Distribucion de ventas**

## País

```python
ventas_por_pais = df.groupby('COUNTRY')['AMOUNT'].sum()

print(ventas_por_pais)

ventas_por_pais = df.groupby('COUNTRY')['AMOUNT'].sum().reset_index()

fig_pais = px.bar(ventas_por_pais, x='COUNTRY', y='AMOUNT', title='Distribución de Ventas por País',
                  labels={'COUNTRY': 'País', 'AMOUNT': 'Ventas (AMOUNT)'})
fig_pais.show()
```

La gráfica destaca una fuerte concentración de ventas en Grán Bretaña, seguida por un desempeño medio en Dinamarca y Holanda, y un volumen bajo de ventas en otros países como España y Portugal.

 Esta distribución desigual sugiere oportunidades de análisis para comprender los factores que contribuyen a estas variaciones y optimizar las estrategias de negocio en consecuencia.

## Año y mes

```python
ventas_por_mes_ano = df.groupby(['YEAR', 'MONTH'])['AMOUNT'].sum().unstack()
print(ventas_por_mes_ano)

ventas_por_mes_ano = df.groupby(['YEAR', 'MONTH'])['AMOUNT'].sum().reset_index()

fig_mes_ano = px.bar(ventas_por_mes_ano, x='MONTH', y='AMOUNT', color='YEAR', barmode='group',
                     title='Distribución de Ventas por Mes y Año',
                     labels={'MONTH': 'Mes', 'AMOUNT': 'Ventas (AMOUNT)', 'YEAR': 'Año'})
fig_mes_ano.show()


```

Analizando la grafica podemos ver una tendencia de crecimiento en las ventas, se observa un claro incremento en las ventas desde 2023 hasta 2025. El año 2023 presenta las ventas más bajas en comparación, mientras que 2024 muestra un aumento considerable.
En 2025, las ventas parecen alcanzar su punto más alto en los meses representados, lo que sugiere un crecimiento continuo en la demanda o en la efectividad de estrategias comerciales.

Aunque la tendencia general es de crecimiento, hay variaciones en las ventas mes a mes. Esto indica que la demanda no es homogénea a lo largo del año y puede estar influenciada por factores externos, como temporadas de vacaciones, festividades o cambios en las condiciones del mercado.

## Marca

```python
ventas_por_marca = df.groupby('SUBBRAND')['AMOUNT'].sum()
print(ventas_por_marca)

import plotly.express as px

ventas_por_marca = df.groupby('SUBBRAND')['AMOUNT'].sum().reset_index()

fig = px.bar(ventas_por_marca, x='SUBBRAND', y='AMOUNT', title='Distribución de Ventas por Submarca',
             labels={'SUBBRAND': 'Submarca', 'AMOUNT': 'Ventas (AMOUNT)'})

fig.show()
```

# **2. Tendencia y estacionalidad**

## Todas las ventas del país con menos ventas

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import plotly.graph_objects as go

pais_menos_ventas = df.groupby('COUNTRY')['AMOUNT'].sum().idxmin()
print(f"País con menos ventas: {pais_menos_ventas}")

df_pais_menos_ventas = df[df['COUNTRY'] == pais_menos_ventas]

ventas_mensuales = df_pais_menos_ventas.groupby(['YEAR', 'MONTH'])['AMOUNT'].sum()
ventas_mensuales.index = pd.to_datetime(ventas_mensuales.index.map(lambda x: f"{x[0]}-{x[1]:02d}"))

descomposicion = seasonal_decompose(ventas_mensuales, model='additive', period=12)

fig = go.Figure()

# Ventas Mensuales
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=ventas_mensuales, mode='lines', name='Ventas Mensuales'))

# Tendencia
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=descomposicion.trend, mode='lines', name='Tendencia'))

# Estacionalidad
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=descomposicion.seasonal, mode='lines', name='Estacionalidad'))

# Residuo
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=descomposicion.resid, mode='lines', name='Residuo'))

# Configuración del diseño del gráfico
fig.update_layout(
    title=f"Descomposición de Ventas para el País con Menos Ventas: {pais_menos_ventas}",
    xaxis_title="Fecha",
    yaxis_title="Ventas (AMOUNT)",
    height=800,
)

# Mostrar el gráfico
fig.show()
```

La gráfica muestra un comportamiento de ventas que crece significativamente durante el primer año y comienza a decaer a mediados de 2024. La tendencia refleja este comportamiento a gran escala, mientras que la estacionalidad muestra la influencia de factores cíclicos predecibles a lo largo del año.

## Marca con más ventas

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import plotly.graph_objects as go

#submarca con más ventas
submarca_mas_ventas = df.groupby('SUBBRAND')['AMOUNT'].sum().idxmax()

#datos para la submarca con más ventas
df_submarca_mas_ventas = df[df['SUBBRAND'] == submarca_mas_ventas]

# Agrupar por año y mes para obtener una serie temporal de ventas mensuales
ventas_mensuales = df_submarca_mas_ventas.groupby(['YEAR', 'MONTH'])['AMOUNT'].sum()
ventas_mensuales.index = pd.to_datetime(ventas_mensuales.index.map(lambda x: f"{x[0]}-{x[1]:02d}"))

# Descomponer la serie temporal
descomposicion = seasonal_decompose(ventas_mensuales, model='additive', period=12)

fig = go.Figure()

# Ventas Mensuales
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=ventas_mensuales, mode='lines', name='Ventas Mensuales'))

# Tendencia
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=descomposicion.trend, mode='lines', name='Tendencia'))

# Estacionalidad
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=descomposicion.seasonal, mode='lines', name='Estacionalidad'))

# Residuo
fig.add_trace(go.Scatter(x=ventas_mensuales.index, y=descomposicion.resid, mode='lines', name='Residuo'))

# Configuración del diseño del gráfico
fig.update_layout(
    title=f"Descomposición de Ventas para la Submarca con Más Ventas: {submarca_mas_ventas}",
    xaxis_title="Fecha",
    yaxis_title="Ventas (AMOUNT)",
    height=800,
)

# Mostrar el gráfico
fig.show()
```

La descomposición muestra que las ventas de Pepsi Max tuvieron un crecimiento sostenido hasta mediados de 2024, cuando alcanzaron un pico, seguido de una desaceleración en las ventas. La tendencia descendente a partir de ese punto podría reflejar una saturación del mercado o un cambio en las preferencias de los consumidores.

# **3. Cuales son las predicciones hechas en España y como de buenas son**



*   AI_forecast  ->  venta que se ha predicho
*   actual  -> venta que se ha producido




```python
import pandas as pd
import plotly.graph_objects as go

data_spain = df[(df['COUNTRY'] == 'Spain') &
                ((df['YEAR'] == 2023) | (df['YEAR'] == 2024)) &
                (df['MONTH'] <= 8 if df['YEAR'].max() == 2024 else True)]

df_actual = data_spain[data_spain['SCENARIO'] == 'actual'].copy()
df_actual['DATE'] = pd.to_datetime(df_actual[['YEAR', 'MONTH']].assign(DAY=1))

sales_actual = df_actual.groupby(df_actual['DATE'].dt.to_period("M"))['AMOUNT'].sum().reset_index()
sales_actual['DATE'] = sales_actual['DATE'].dt.to_timestamp()

forecasts = [f'AI_P{str(i).zfill(2)}F' for i in range(2, 12)]

# Crear una gráfica para cada forecast
for forecast in forecasts:
    # Filtrar los datos de predicción para el forecast actual
    df_forecast = data_spain[(data_spain['SCENARIO'] == 'AI_forecast') & (data_spain['FORECAST'] == forecast)].copy()
    df_forecast['DATE'] = pd.to_datetime(df_forecast[['YEAR', 'MONTH']].assign(DAY=1))

    sales_forecast = df_forecast.groupby(df_forecast['DATE'].dt.to_period("M"))['AMOUNT'].sum().reset_index()
    sales_forecast['DATE'] = sales_forecast['DATE'].dt.to_timestamp()

    # Unir ambos DataFrames por fecha, manteniendo solo las filas donde hay datos en ambos
    comparacion = pd.merge(sales_actual, sales_forecast, on='DATE', how='inner', suffixes=('_Actual', '_Forecast'))
    comparacion = comparacion.dropna(subset=['AMOUNT_Forecast'])

    # Calcular el MAE entre ventas reales y predichas
    comparacion['MAE'] = abs(comparacion['AMOUNT_Actual'] - comparacion['AMOUNT_Forecast'])

    fig = go.Figure()

    # Añadir la línea de ventas reales
    fig.add_trace(go.Scatter(
        x=comparacion['DATE'],
        y=comparacion['AMOUNT_Actual'],
        mode='lines+markers',
        name='Ventas Reales (Actual)',
        line=dict(color='blue')
    ))

    # Añadir la línea de ventas predichas
    fig.add_trace(go.Scatter(
        x=comparacion['DATE'],
        y=comparacion['AMOUNT_Forecast'],
        mode='lines+markers',
        name=f'Ventas Predichas ({forecast})',
        line=dict(color='red')
    ))

    #MAE
    fig.add_trace(go.Bar(
        x=comparacion['DATE'],
        y=comparacion['MAE'],
        name=f'MAE: {comparacion["MAE"].mean():.2f}',
        marker_color='gray',
        opacity=0.3
    ))

    fig.update_layout(
        title=f'Comparación de Ventas Reales vs. Predichas ({forecast}) en España',
        xaxis_title='Fecha',
        yaxis_title='Cantidad Ventas',
        legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5),
        template="plotly_white"
    )

    fig.show()
```
